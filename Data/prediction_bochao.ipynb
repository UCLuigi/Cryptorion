```python
for i in range (3,34,3):
    five_min_later=df1.loc[12:(len(df2)+11),[l[i]]].apply(pd.to_numeric)
    five_min_before=df1.loc[11:(len(df2)+10),[l[i]]].apply(pd.to_numeric)
    five_min_later.index=five_min_before.index=range(0,len(df2))
    
    #print (type(five_min_before))
    #print(type(five_min_later))
    trend=five_min_later-five_min_before
    
    #price increase 1
    #price decrease -1
    #print stagnent 0
    df2[l[i]+'trend in 5min']=trend
    df2.loc[(df2[l[i]+'trend in 5min']==0),l[i]+'trend in 5min']=0
    df2.loc[(df2[l[i]+'trend in 5min']<0),l[i]+'trend in 5min']=-1
    df2.loc[(df2[l[i]+'trend in 5min']>0),l[i]+'trend in 5min']=1

df2.head()
```

```python
df3=pd.read_csv('pretreated 20000 data points starting from 2017-01-01')
df3.head()
```

```python
#print(df3.iloc[0:len(df2)-200,1:408].head())
x_train=df3.iloc[0:len(df3)-200,0:407].apply(pd.to_numeric)
print(len(df3))
x_test=df3.iloc[len(df3)-200:len(df3),0:407].apply(pd.to_numeric)

print(len(df2))
y_train=df2.loc[0:len(df2)-201,'  USDT_BTC.2\n0      pricetrend in 5min'].apply(pd.to_numeric)
y_test=df2.loc[len(df2)-200:len(df2),'  USDT_BTC.2\n0      pricetrend in 5min'].apply(pd.to_numeric)
#print(x_train)
print('x_train', len(x_train))
print('y_train', len(y_train))
print('x_test', len(x_test))
print('y_test', len(y_test))
```

```python
LogReg=LogisticRegression()
LogReg.fit(x_train,y_train)
LogReg_training_accu=LogReg.score(x_train,y_train)
LogReg_test_accu=LogReg.score(x_test,y_test)

print('LogReg Training accuracy:',LogReg_training_accu,'\n','LogReg Test accuracy:',LogReg_test_accu,'\n')
```

```python
svc = SVC()
svc.fit(x_train, y_train)
#Y_pred = svc.predict(x_db_test)
svc_training_accu = svc.score(x_train, y_train)
svc_test_accu=svc.score(x_test, y_test)
print ('SVC Training accuracy:',svc_training_accu, '\n','SVC Test accuracy:',svc_test_accu,'\n')

```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train, y_train)
#Y_pred = knn.predict(X_val)
knn_training_accu = knn.score(x_train, y_train)
knn_test_accu=knn.score(x_test, y_test)
print ('KNN Training accuracy:',knn_training_accu, '\n','KNN Test accuracy:',knn_test_accu,'\n')
```

```python

perceptron = Perceptron(max_iter=1000)
perceptron.fit(x_train,y_train)
#Y_pred = perceptron.predict(X_val)
perce_training_accu= perceptron.score(x_train, y_train) 
perce_test_accu=perceptron.score(x_test, y_test) 
print ('Perceptron Training accuracy:',perce_training_accu, '\n','Perceptron Test accuracy:',perce_test_accu,'\n')
```

```python

gradboost = xgb.XGBClassifier(n_estimators=1000)
gradboost.fit(x_train,y_train)
#Y_pred = gradboost.predict(x_db_train)
xgb_training_accu = gradboost.score(x_train,y_train)
xgb_test_accu=gradboost.score(x_test, y_test)
print ('xgboost Training accuracy:',xgb_training_accu, '\n','xgboost Test accuracy:',xgb_test_accu,'\n')


```

```python
random_forest = RandomForestClassifier(n_estimators=1000)
random_forest.fit(x_train,y_train)
#Y_pred = random_forest.predict(X_val)
forest_training_accu = random_forest.score(x_train,y_train)
forest_test_accu=random_forest.score(x_test, y_test)
print ('random forest Training accuracy:',forest_training_accu, '\n','random forest Test accuracy:',forest_test_accu,'\n')
```
