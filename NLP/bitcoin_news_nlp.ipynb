```python
from newsapi import NewsApiClient
from bs4 import BeautifulSoup
import requests
import pandas as pd

newsapi = NewsApiClient(api_key='e6ab46ae97e04bf0b1180daf707c8b8c')

# top_headlines = newsapi.get_top_headlines(q='bitcoin', language='en')
news = newsapi.get_everything(q='bitcoin', sort_by='popularity', page_size=100)
data = []

articles = news["articles"]
for article in articles:
    link = article["url"]
    date = article['publishedAt'][:10]
    r = requests.get(link)
    bs = BeautifulSoup(r.text, 'html.parser')
    if bs.find('title') == None:
        continue
    article_title = bs.find('title').text
    
    article_words = ""
    for paragraph in bs.find_all('p'):
        if paragraph.text == 'Advertisement':
            continue
        #sentences = paragraph.text.split(". ")
        #print(sentences)
        article_words += paragraph.text
    data.append([date, article_title, article_words])
    
news_articles = pd.DataFrame(data, columns=['Date', 'Title', 'article'])    
news_articles        
```
