```python
import pandas as pd
import datetime
import ipywidgets as widgets
```

```python
from IPython.display import display
w = widgets.Dropdown(
    options={'BTC/USDT':"USDT_BTC", 'ETH/BTC':"BTC_ETH", 'LTC/BTC':"BTC_LTC", 'XRP/BTC':"BTC_XRP"},
    description='Currency Pair:',
)
display(w)
```

```python
days = int(input("How many days (0 for all data) : "))
freq = int(input("Frequency (in seconds) : "))

def retrieve(days = days, freq = freq, pairs = [w.value]):
    df_list = list()
    period = days and (datetime.datetime.now() - datetime.timedelta(days)).timestamp()
    print('\nCurrency pairs: ', ', '.join(pairs), '({} pairs)'.format(len(pairs)))
    f = widgets.FloatProgress(min=0, max=len(pairs))
    print('\nProgress: '), display(f)
    
    for p in pairs: 
        url = "https://poloniex.com/public?command=returnChartData&currencyPair={}&start={}&end=9999999999&period={}".format(p, period, freq)
        f.value += 0.1
        temp = pd.read_json(url).set_index('date')
        data = dict()
        f.value += 0.4
        for k,v in temp.to_dict().items():
            data[(p, k)] = v
        df_list.append(pd.DataFrame(data))
        f.value += 0.5
    
    print("done!")
    return df_list


pairs = ['USDT_BTC']#, 'USDT_XRP', 'USDT_BCH', 'USDT_ETH', 'USDT_LTC', 'USDT_ETC', 'USDT_STR', 'USDT_XMR', 'USDT_ZEC', 'USDT_DASH', 'USDT_NXT', 'USDT_REP']

data = retrieve(pairs=pairs)
data[0].head()
```

```python
for idx, df in enumerate(data):
    data[idx] = df.loc[df.index >= '2017', (slice(None), ['weightedAverage', 'quoteVolume', 'volume'])].rename(columns={'weightedAverage':'price'}, level=1)
    display(data[idx].head())
    
```

```python
df = pd.concat(data, axis=1)
df.head()
```

```python
df.to_csv("./Data/poloniex_usdt_pairs30min.csv")
```
